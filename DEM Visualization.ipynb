```python
import rasterio
import matplotlib.pyplot as plt
import os
```

```python
# Function to load DEM data from a .tif file
def load_dem(filename):
    with rasterio.open(filename) as dem_file:
        dem_data = dem_file.read(1)  # Read the first (and typically only) band of data
    return dem_data

# Function to visualize DEM data
def visualize_dem(dem_data):
    plt.figure(figsize=(10, 6))
    plt.imshow(dem_data, cmap='terrain')  # 'terrain' colormap gives a natural look to the DEM
    plt.colorbar(label='Elevation')
    plt.title('Digital Elevation Model (DEM)')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.show()
```

```python
# Load and visualize the DEM
hillshade_file = 'CTX_DEM/B20hillshade.tif'  # Replace with the path to  .tif file
hillshade_data = load_dem(hillshade_file)
visualize_dem(hillshade_data)

#dtm_file = 'CTX_DEM/B20dtm.tif'  # Replace with the path to  .tif file
#dtm_data = load_dem(dtm_file)
#visualize_dem(dtm_data)
```

```python
dem_filename = 'MOLA-HRSC Blended/Mars_HRSC_MOLA_BlendDEM_Global_200mp_v2.tif'  # Replace with the path to  .tif file
dem_data = load_dem(dem_filename)
visualize_dem(dem_data)
```
