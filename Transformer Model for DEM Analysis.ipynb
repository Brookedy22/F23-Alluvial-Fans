```python
#Load and Visualize DEM Data: You'll need to load the DEM data from the .tif file, and possibly preprocess it (normalization, scaling, etc.).

import rasterio
import numpy as np
import matplotlib.pyplot as plt

def load_dem(filename):
    with rasterio.open(filename) as dataset:
        dem = dataset.read(1)
    return dem
    
# Function to visualize DEM data
def visualize_dem(dem_data):
    plt.figure(figsize=(10, 6))
    plt.imshow(dem_data, cmap='gray')  # 'terrain' colormap gives a natural look to the DEM, using 'gray' for hillshade
    plt.colorbar(label='Hillshade Direction')
    plt.title('Hillshade')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.show()

dem_data = load_dem("CTX_DEM/B20hillshade.tif")
visualize_dem(dem_data)
```

```python
from rasterio.plot import show
from scipy.ndimage import generic_filter

def calculate_slope(dem, cell_size):
    # Define a function to calculate slope in the 3x3 window
    def slope(window):
        dzdx = ((window[2] + 2 * window[5] + window[8]) - (window[0] + 2 * window[3] + window[6])) / (8 * cell_size)
        dzdy = ((window[6] + 2 * window[7] + window[8]) - (window[0] + 2 * window[1] + window[2])) / (8 * cell_size)
        return np.sqrt(dzdx**2 + dzdy**2) * 100  # Convert to percentage

    # Apply the slope function on each cell
    slopes = generic_filter(dem, slope, size=3)
    return slopes

def categorize_slopes(slopes, thresholds):
    categories = np.digitize(slopes, thresholds)
    return categories

def main():
    dem = load_dem("CTX_DEM/B20hillshade.tif")

    # Assuming a cell size in meters, adjust as per your DEM's resolution
    cell_size = 1 

    slopes = calculate_slope(dem, cell_size)

    # Define your thresholds for slope categorization
    # Example: [10, 20] means three categories: <10%, 10-20%, >20%
    thresholds = [2,8.5] 

    slope_categories = categorize_slopes(slopes, thresholds)

    # Display the Slope and Categorized Slopes (optional)
    show(slopes, title='Slope')
    show(slope_categories, title='Categorized Slopes')

if __name__ == "__main__":
    main()
```
